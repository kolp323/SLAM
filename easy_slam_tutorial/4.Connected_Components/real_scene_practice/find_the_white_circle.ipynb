```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
circles_img_mat = cv2.imread('circles.jpg', cv2.IMREAD_GRAYSCALE)
# 二值化操作
threshold, circles_img_mat = cv2.threshold(
    src=circles_img_mat, 
    thresh=0, 
    maxval=255, 
    type= cv2.THRESH_BINARY + cv2.THRESH_OTSU
)

# 识别连通域
ret, labels = cv2.connectedComponents(circles_img_mat)
# 用矩形框标记所识别的连通域（除了0号背景）
for i in range(1, ret):
    y, x = np.where(labels==i)
    labeled_img_mat = cv2.rectangle(
        circles_img_mat,
        pt1=(x.min(), y.min()),
        pt2=(x.max(), y.max()),
        color=65,
        thickness=3
    )

plt.imshow(labeled_img_mat)
plt.show()
```
